```python
# Hand writting digit prediction pixel size 28*28
# Normilization 
# Learning algorithm  - Adam - for intelligent learning value adoption at runtime
# Activation - ReLU - best for hidden layers as max(0,z)
# Output layer activation - Softmax
# Sigmoid curve is flat at two values so learning will be slow
# Loss - Sparse categorical  - generilization of binnary classification
# Tensorfolw expect feature row wise single row represent a complete feature

import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train[0].shape
print(x_train.shape)
print(y_train.shape)
```

```python
plt.imshow(x_train[20], cmap= plt.cm.binary)
plt.show()
```

```python
x_train = tf.keras.utils.normalize(x_train , axis=1)
x_test  = tf.keras.utils.normalize(x_test , axis=1)
plt.imshow(x_train[20], cmap= plt.cm.binary)
plt.show()
```

```python
model = Sequential(
        [
            tf.keras.layers.Flatten(),
            Dense(units = 25, activation = 'relu'),
            Dense(units = 15, activation = 'relu'),
            Dense(units = 10,  activation = 'linear')
        ]
)
```

```python
model.compile(optimizer =tf.keras.optimizers.Adam(0.001) ,loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))
```

```python
model.fit(x_train, y_train, epochs=5)
```

```python
 model.evaluate(x_test, y_test)
```

```python
# Evaluate the test dataset
predict = model.predict(x_test)
print("Evaluation of test set")
for i in range (predict.shape[0]):
    print (np.argmax(predict[i]) , end=' ')
```
