```python
# Fasion Mnist
import tensorflow as tf
import numpy as np
import tensorflow.keras.datasets.fashion_mnist as fm
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# 0	T-shirt/top
# 1	Trouser
# 2	Pullover
# 3	Dress
# 4	Coat
# 5	Sandal
# 6	Shirt
# 7	Sneaker
# 8	Bag
# 9	Ankle boot
```

```python
(x_train, y_train), (x_test, y_test) = fm.load_data()
```

```python
x_train.shape[1]
```

```python
plt.imshow(x_train[6], cmap= plt.cm.binary)
plt.show()
plt.imshow(x_train[6])
plt.show()
```

```python
x_train = tf.keras.utils.normalize(x_train , axis=1)
x_test = tf.keras.utils.normalize(x_test , axis=1)
```

```python
plt.imshow(x_train[20], cmap= plt.cm.binary)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(),
    Dense(units = 28 , activation = 'relu'),
    Dense(units = 28 , activation = 'relu'),
    Dense(units = 10 , activation = 'sigmoid')
])
```

```python
model.compile(optimize = tf.keras.optimizers.Adam(.001), loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))
```

```python
model.fit(x_train,y_train, epochs=5)
```

```python
predict = model.predict(x_test[0].reshape(1, 28, 28))
np.argmax(predict)
```
